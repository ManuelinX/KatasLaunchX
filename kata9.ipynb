```python
def informe2(horaPreDespegue,Destino,*tiempo,**tanques):
    return f"""
    La hora de despegue fue a las {horaPreDespegue}
    El destino es {Destino}
    El tiempo de vuelo es de {sum(tiempo)} minutos
    El total de combustible son {sum(tanques.values())}
    """


print(informe2(10,'Marte',30,80,99,externo = 30,interno = 55))



def informe3(horaPreDespegue,Destino,*tiempo,**tanques):
    inf = f"""
    La hora de despegue fue a las {horaPreDespegue}
    El destino es {Destino}
    El tiempo de vuelo es de {sum(tiempo)} minutos"""

    for key, gas in tanques.items():
 
        inf += f"""\n    El tanque {key} tiene {gas} galones"""
    
    totalGas = f"""
    El total de combustible es {sum(tanques.values())}
    """

    inf += totalGas

    return inf

print(informe3(10,'Marte',30,30,40,interno = 90,externo = 55))

```
