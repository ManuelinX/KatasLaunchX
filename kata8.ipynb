```python
from math import floor

planet_moons = {
    'mercury': 0,
    'venus': 0,
    'earth': 1,
    'mars': 2,
    'jupiter': 79,
    'saturn': 82,
    'uranus': 27,
    'neptune': 14,
    'pluto': 5,
    'haumea': 2,
    'makemake': 1,
    'eris': 1
}

planets = len(planet_moons)
moons = planet_moons.values()

total_moons = 0

for i in moons:
    total_moons += i

promedio = total_moons/planets
promedio = floor(promedio)

print(f'El promedio de lunas por planeta son: {promedio} lunas')
```
